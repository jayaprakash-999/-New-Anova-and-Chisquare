# **Chisquare**

Checks if there's a relationship between categorical variables

```python
import pandas as pd
import numpy as np


df = pd.read_csv("mtcars.csv")
```

```python
df.head()
```

```python
crosstable = pd.crosstab(df['cyl'],df['gear'])
```

```python
crosstable
```

```python
from scipy.stats import chi2_contingency
chi, p, dof, expected = chi2_contingency(crosstable)


# imports the chi2_contingency function from the scipy.stats module
# Chi chi square.

# p  The p-value,  the probability of observing the data (or more extreme data) if there were no relationship between the variables
# A low p-value (typically less than 0.05) suggests that there is a statistically significant relationship.

# degree of fredem 3*3, =3-1 *3-1 =4
# expected: The expected frequencies.
```

```python
chi
```

```python
p

# p value < Alpha
# 0.0012  < 0.05
# Reject the H0

# p value < Alpha                            --> Reject the H0
# p value > Alpha                            --> Accept the H0
# X^2 calculated < X^2 critical or table     --> Accept the H0


# 'cyl','gear' are Dependent or relationship b/w them
```
